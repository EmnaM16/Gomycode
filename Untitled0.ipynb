```python
def last(n): return n[-1]

def sort_list_last(tuples):
  return sorted(tuples, key=last)

print(sort_list_last([(2, 5), (1, 2), (4, 4), (2, 3), (2, 1)]))
```

```python
d1={'a': 100, 'b': 200, 'c':300}
d2= {'a': 300, 'b': 200, 'd':400}
d3={}
for key in d2:
  if key in d1:
    d3[key]=d2[key]+d1[key]
  else:
    d3[key] = d2[key]

for key in d1:
  if key not in d2:

    d3[key] = d1[key]
print(d3)
```
